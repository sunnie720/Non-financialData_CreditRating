```python
special_char='※\"' # 특수문자 제거
for char in special_char:
    report_text=[w.replace(char, '') for w in report_text]

report_text=''.join(report_text)
print(report_text)
```

```python
import konlpy
from konlpy.tag import Kkma, Komoran, Hannanum, Okt
from konlpy.utils import pprint
from nltk import Text
```

```python
# 형태소 확인
komoran=Komoran()
komoran_morphs=komoran.pos(report_text)
print(komoran_morphs)
```

```python
# 명사만 확인
nouns=komoran.nouns(report_text)
print(nouns)
print('명사 개수:', len(nouns))
```

```python
# 2자 이상인 단어들만 추출
nouns=[noun for noun in nouns if len(noun)>1]
print('명사 개수:', len(nouns))
```

## 빈도 분석

```python
from collections import Counter

c=Counter(nouns)
print(c)
print('중복 제거 개수:', len(c))
```

```python
c.most_common(20)
```

```python
# 단어 빈도가 3번 이하인 단어 제거
c_fq={}
for k,v in c.items():
  if c[k]>3:
    c_fq[k]=v
print(c_fq)
print('빈도 3 초과 명사 개수:', len(c_fq))
```

## 불용어 제거

```python
inwords=list(c_fq.keys())
stopwords=['당사', '본사', '부문', '산하', '총괄', '생산', '판매', '법인', '기업', '구성', '사업', '제품', '사항', 
           '참조', '운영', '사업장', '담당', '활동', '포함', '2021년', '매출', '전년', '대비', '상황', '비중', '기준', 
           '시간', '산출', '실제', '자산', '장부', '금액', '기타', '예정', '업체', '시장', '전체', '수준', '영향', '재무', 
           '요인', '한편', '권역', '센터', '발생', '평가', '주기', '표시', '경우', '반영', '다음', '목적', '만원', '해당', 
           '관련', '상호', '사용', '업계', '산업', '조직', '분야', '기반', '적용', '경영', '이후', '19', '인해', '대로', 
           '출처', '추이', '자료', '제품군', '전망']

final_nouns=[]
for i in range(len(nouns)):
  if nouns[i] in inwords:
    if nouns[i] not in stopwords:
      final_nouns.append(nouns[i])
print(final_nouns)
print('최종 명사 단어 개수:', len(final_nouns))  
```

```python
# 자연어처리 라이브러리 설치
!git clone https://github.com/lovit/textrank.git
!pip install pororo
!pip install konlpy
```

```python
import textrank
from textrank.textrank import KeywordSummarizer
```

```python
# 명사(NN), 어근(XR), 형용사(VA), 동사(VV)만 추출하는 함수
def komoran_tokenize(sent):
    words = komoran.pos(sent, join=True)
    words = [w for w in words if ('/NN' in w or '/XR' in w or '/VA' in w or '/VV' in w)]
    return words
```

```python
sents = report_text.split('다.')
sents = [sent + '다.' for sent in sents]
sents[:5]
```

- 상위 키워드 추출

```python
keyword_extractor=KeywordSummarizer(
    tokenize=komoran_tokenize,
    min_count=2,
    window=-1,                     # cooccurrence within a sentence
    min_cooccurrence=2,
    vocab_to_idx=None,             # you can specify vocabulary to build word graph
    df=0.85,                       # PageRank damping factor
    max_iter=30,                   # PageRank maximum iteration
    verbose=False
)
```

```python
keywords=keyword_extractor.summarize(sents, topk=500)
keywords
```

```python
# 상위500개 단어 중 두 글자 이상인 단어(keywords2)
keywords = keyword_extractor.summarize(sents,topk=500)
keywords2=[]
for word, rank in keywords:
  if len(word.split('/')[0])>=2: 
    keywords2.append((word,rank))
keywords2
```

```python
finding='위험'
matching=[s for s in keywords2 if finding in s[0]]
matching_idx=[i for i in range(len(keywords2)) if finding in keywords2[i][0]]
print(matching)
print(matching_idx)
```
